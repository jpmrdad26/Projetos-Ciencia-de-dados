```python
#Vamos organizar e entender o impacto que essas ausencias tem em cada coluna. 

def missing_values_table(df_2):
    # Conta os valores nulos e o percentual de valores nulos
    mis_val = df_2.isnull().sum()
    mis_val_percent = 100 * df_2.isnull().sum() / len(df_2)
    
    # Filtra colunas que possuem valores nulos
    cols_com_nulos = df_2.columns[df_2.isnull().any()]
    
    # Calcula as estatísticas descritivas apenas para colunas com valores nulos
    media = df_2[cols_com_nulos].mean()
    mediana = df_2[cols_com_nulos].median()
    moda = df_2[cols_com_nulos].mode().iloc[0]  # Seleciona a primeira moda (caso haja várias)
    maximo = df_2[cols_com_nulos].max()
    minimo = df_2[cols_com_nulos].min()
    
    # Cria uma tabela combinando todas as informações
    mis_val_table = pd.concat([mis_val[cols_com_nulos], 
                               mis_val_percent[cols_com_nulos], 
                               media, mediana, moda, maximo, minimo], axis=1)
    
    # Renomeia as colunas
    mis_val_table.columns = ['Missing Values', '% of Total Values', 'Mean', 'Median', 'Mode', 'Max', 'Min']
    
    # Filtra colunas que possuem valores nulos
    mis_val_table_ren_columns = mis_val_table.sort_values('% of Total Values', ascending=False).round(1)
    
    # Exibe a quantidade de colunas com valores nulos
    print("O dataframe possui " + str(df_2.shape[1]) + " colunas.\n"      
          "Sendo que " + str(mis_val_table_ren_columns.shape[0]) + " colunas possuem valores nulos.")
    
    return mis_val_table_ren_columns
```

### 2.3.2. Outliers <a class="anchor" id="Outliers"></a>
Na primeira analise, também observamos que havia algumas diferenças relevantes entre as estatisticas descritivas (media, mediana, max,etc), isso pode ser um indicativo de outliers,, vamos analisar as colunas para entender melhor se devemos manter ou remover os valores.  Para ser considerado um outilier deve:

- Abaixo de : 1º quartile − 3 ∗ Intervalo Interquartilico
- Acima de : 3º quartile + 3 ∗ Intervalo Interquartilico

https://www.itl.nist.gov/div898/handbook/prc/section1/prc16.htm



```python
# Selecionar apenas as colunas numéricas do DataFrame
numeric_columns = df_2.select_dtypes(include='number').columns

# Definir o número de colunas e linhas para os subplots
ncols = 2
nrows = (len(numeric_columns) + ncols - 1) // ncols  # Número de linhas necessário

# Definir o tamanho da figura
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 5, nrows * 5))

# Ajustar para o caso em que há apenas uma linha ou uma coluna
axes = axes.flatten()

# Loop para criar um boxplot para cada coluna numérica
for i, column in enumerate(numeric_columns):
    sns.boxplot(y=df_2[column], ax=axes[i])
    axes[i].set_title(f'Boxplot de {column}')

# Se houver subplots adicionais que não foram utilizados, removê-los
for j in range(len(numeric_columns), len(axes)):
    fig.delaxes(axes[j])

# Ajusta o layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

```python

# Função para detectar outliers e calcular contagem e percentual por coluna
def contar_outliers(df_2):
    outlier_counts = {}  # Dicionário para armazenar os resultados

    total_rows = len(df_2)  # Total de linhas no DataFrame

    for column in df_2.select_dtypes(include='number').columns:
        # Calcular o IQR
        Q1 = df_2[column].quantile(0.25)
        Q3 = df_2[column].quantile(0.75)
        IQR = Q3 - Q1

        # Definir limites para outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Contar os outliers
        num_outliers = ((df_2[column] < lower_bound) | (df_2[column] > upper_bound)).sum()
        
        # Calcular o percentual de outliers
        perc_outliers = (num_outliers / total_rows) * 100
        
        # Armazenar os resultados
        outlier_counts[column] = {'Número de Outliers': num_outliers, 'Percentual de Outliers (%)': perc_outliers}

    # Converter o dicionário em um DataFrame
    outliers_df = pd.DataFrame.from_dict(outlier_counts, orient='index').reset_index()
    outliers_df.columns = ['Coluna', 'Número de Outliers', 'Percentual de Outliers (%)']
    
    return outliers_df

# Aplicar a função ao DataFrame
outliers_df = contar_outliers(df_2)
outliers_df

```

```python
# Substitua 'X' pelo nome da coluna que você deseja analisar
coluna_x = 'Life expectancy '

# Calcula a matriz de correlação
correlacoes = df_3.corr(numeric_only=True)

# Filtra apenas as correlações envolvendo a coluna X
correlacoes_x = correlacoes[[coluna_x]].drop(coluna_x)

# Filtra correlações fortes
correlacoes_fortes_positivas = correlacoes_x[correlacoes_x[coluna_x] > 0.35]
correlacoes_fortes_negativas = correlacoes_x[correlacoes_x[coluna_x] < -0.35]

# Lista para armazenar pares de variáveis correlacionadas
pares_correlatos = []

# Adiciona pares de correlações fortes positivas
for idx in correlacoes_fortes_positivas.index:
    if pd.notna(correlacoes_fortes_positivas.loc[idx, coluna_x]):
        pares_correlatos.append((coluna_x, idx, correlacoes_fortes_positivas.loc[idx, coluna_x]))

# Adiciona pares de correlações fortes negativas
for idx in correlacoes_fortes_negativas.index:
    if pd.notna(correlacoes_fortes_negativas.loc[idx, coluna_x]):
        pares_correlatos.append((coluna_x, idx, correlacoes_fortes_negativas.loc[idx, coluna_x]))

# Retirar duplicidade e criar um dicionário
pares_unicos = set()
dict_correlacoes = {}

for par in pares_correlatos:
    par_ordenado = tuple(sorted(par[:2]))
    correlacao = par[2]
    pares_unicos.add(par_ordenado)
    dict_correlacoes[par_ordenado] = correlacao

# Convertendo para lista
correlacoes_unicas_lista = [(col1, col2, correlacao) for (col1, col2), correlacao in dict_correlacoes.items()]

# Exibir resultado
correlacoes_unicas_lista
```
